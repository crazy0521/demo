```python
#Importing libraries
import pandas as pd
import numpy as np



#a)Loading the dataset
dataset = pd.read_csv('Heart (1).csv')
dataset.head(10)



#b)Finding the shape of the dataset
dataset.shape



#c)Finding missing values
dataset.isnull().sum()



#c)Filling the missing values of dataset with mean - Cleaning the data
dataset.fillna(dataset.mean)



#d)Finding data type of each column
dataset.info()



#e)Finding out total zeros
(dataset == 0).sum().sum()



#e)Finding out number of zeros column wise
(dataset == 0).sum()



#f)Indexing data
#First tuple gets displayed
dataset.iloc[0]



#f)Indexing data using negative indexing
#Last tuple gets displayed
dataset.iloc[-1]



#f)Indexing data in a range
dataset.iloc[5:10]  #selects rows 2 to 4



#f)Selecting data - Columns
dataset["Chol"]



#f)Selecting data - Multiple Columns
dataset[["Age", "Chol"]]



#f)Sorting data
sorted_dataset = dataset.sort_values(by='Age', ascending=True)
sorted_dataset.head()



#g)Describing attributes of data and Data types of each column
dataset.info()
dataset.describe()



#h)Counting unique values of data,, converting variable data type (e.g. from long to short, vice versa)
unique_values = dataset["Age"].value_counts()
print(unique_values)



#h)Displaying format of each column
dataset.info()



#h)Converting variable datatype
#Oldpeak converted into Integer
dataset['Oldpeak'] = dataset['Oldpeak'].astype('int32')
dataset.info()
```
